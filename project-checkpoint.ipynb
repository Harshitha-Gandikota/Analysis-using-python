## Data Visualization

```python
#from the below graph we conclude processor 8 mobiles having high price
plt.bar(data['Processor'],data['Price'],color=['r','g','w','y'])
plt.title('bar graph')
plt.show()
```

```python
#Histplot
se.histplot(data=data, x='Price', kde=True)
plt.title('Price Distribution')
plt.show()
```

```python
#top 10 mostly used mobiles brand
brand=data['Brand'].value_counts()
top5_brands=brand.head(5)
print(top5_brands,end=" ")
```

```python
plt.plot(top5_brands,'Dm',ms=10)
plt.show()
```

```python
mylabels = ["Intex", "Samsung", "Micromax", "Lava","Panasonic"]
plt.pie(top5_brands,labels=mylabels,shadow=True)
plt.legend()
plt.show() 
```

```python
plt.plot(data['Operating system'],data['Price'],'ok')
plt.show()
```

```python
se.distplot(data['Price'])
plt.show()
```

```python
se.kdeplot(data["Resolution x"])
se.kdeplot(data["Resolution y"])
```

```python
# Visualize the price distribution
plt.figure(figsize=(10, 6))
se.histplot(data['Price'], kde=True)
plt.xlabel('Price')
plt.title('Price Distribution')
plt.show()
```

```python
# Scatter plot between price and a specific feature (e.g., RAM)
plt.figure(figsize=(10, 6))
se.scatterplot(x='RAM (MB)', y='Price', data=data)
plt.xlabel('RAM')
plt.ylabel('Price')
plt.title('Price vs. RAM')
plt.show()
```

```python
plt.scatter(data['Screen size (inches)'], data['Price'])
plt.xlabel('Screen size (inches)')
plt.ylabel('Price')
plt.title('Screen Size vs. Price')
plt.show()
```

```python
#box plot
se.boxplot(data=data, x='Internal storage (GB)')
plt.title('Price Distribution')
plt.show()
```

```python
plt.figure(figsize=(10, 15))
plt.plot(data['Price'],data['Brand'], 'o:r')
plt.show()
```
